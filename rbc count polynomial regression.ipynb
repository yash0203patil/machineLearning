```python
#importing the libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the dataset
dataset=pd.read_csv("human rbc counts.csv")
X=dataset.iloc[:,0:-1].values
y=dataset.iloc[: , -1].values
```

```python
#training the linear regression model on the whole dataset
from  sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X,y)
```

```python
#training the whole dataset  for the polynomial linaer regression model
from sklearn.preprocessing import PolynomialFeatures
polynomial_reg =PolynomialFeatures(degree = 4)
X_poly = polynomial_reg.fit_transform(X)
linear_reg_2 = LinearRegression()
linear_reg_2.fit(X_poly,y)
```

```python
#visualizing the linear regression results 
plt.scatter(X , y , color="red")
plt.plot(X, linear_reg.predict(X) , color='blue')
plt.title("RBC - linaer Regression")
plt.xlabel('Age')
plt.ylabel('RBC')
plt.show()
```

```python
#visualizing the polynomial regression 
plt.scatter(X, y , color="red")
plt.plot(X , linear_reg_2.predict(X_poly), color="blue")
plt.title("RBC - polynomial Regression")
plt.xlabel('Age')
plt.ylabel('RBC')
plt.show()
```

```python
##predicting the result with user input values
linear_reg.predict([[6.5]])
```
